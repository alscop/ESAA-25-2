EDA
- bar chart: 요일에 따른 18-20시 탑승객 수
- line graph : 출근시간 승차인원과 퇴근시간 승차인원 -> 출근시간 승차인원 합, 평균과 퇴근시간 승차인원 합, 평균이 같아 버스타고 출근 시 퇴근도 버스타고 한다고 추측 가능.
- ine graph : 출근시간 하차인원과 퇴근시간 승차인원 -> 출근시간 하차인원 합, 평균과 퇴근시간 승차인원 합, 평균이 같아 출근시간 하차인원 많은 곳에서는 퇴근시간 승차인원 많을 것으로 예상.
- ine graph : 오전시간 승/하차인원과 퇴근시간 승/하차인원 : 오전시간에 많다면 그날 유동인구 多. 주중에 많음.
- 분포 그래프: 배차 간격에 따른 퇴근시간 탑승객 수 : 수요 많을수록 배차 간격 짧음, 퇴근시간 탑승객 수 많음
- 분포 그래프: 소비 비율에 따른 퇴근시간 탑승객 수 : 소비액 높을수록 퇴근시간 탑승객 수 많음(소비자, 유동인구 많음)
- bar chart: 제주도의 시별 탑승객 수 -> 제주시가 서귀포시보다 많음
