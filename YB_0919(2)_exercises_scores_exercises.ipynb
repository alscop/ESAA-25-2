````python
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}
df = pd.DataFrame(raw_data)
df
````

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

````python
plt.figure(figsize=(8,6))

plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age'], alpha=0.6)
plt.title("Pre-test Score Vs Post-test Score")
plt.xlabel("Pre-test Score")
plt.ylabel("Post-test Score")

plt.grid(True)
plt.show()
````

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

````python
plt.figure(figsize=(8,6))

plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age']*4.5, c=df['female'], alpha=0.6)
plt.title("Pre-test Score Vs Post-test Score")
plt.xlabel("Pre-test Score")
plt.ylabel("Post-test Score")

plt.grid(True)
plt.show()
````

````python
plt.figure(figsize=(8,6))

# 여성 데이터만 따로 시각화
plt.scatter(df[df['female']==1]['preTestScore'],
            df[df['female']==1]['postTestScore'],
            s=df[df['female']==1]['age']*4.5,
            c='red', # 여성의 색상을 빨간색으로 지정
            alpha=0.6,
            label='Female') # 범례를 위해 라벨 추가

# 남성 데이터만 따로 시각화
plt.scatter(df[df['female']==0]['preTestScore'],
            df[df['female']==0]['postTestScore'],
            s=df[df['female']==0]['age']*4.5,
            c='blue', # 남성의 색상을 파란색으로 지정
            alpha=0.6,
            label='Male') # 범례를 위해 라벨 추가

plt.title("Pre-test Score Vs Post-test Score")
plt.xlabel("Pre-test Score")
plt.ylabel("Post-test Score")
plt.legend() # 범례가 올바르게 표시됨

plt.grid(True)
plt.show()
````

범례 표시를 위해서는 plt.scatter함수 내에 label 지정을 해야하고, 이를 위해선 남성 따로, 여성 따로 그래프를 출력해야함.

사전 점수가 낮았던 학생이 사후 점수가 얼마나 올랐는지 확인하기.

````python
df['score_diff'] = df['postTestScore'] - df['preTestScore']
df.plot(x='first_name', y='score_diff', kind='bar', figsize=(8,6), color='teal')
plt.title('Score Improvement by Student')
plt.xlabel('Student')
plt.ylabel('Score Improvement')
plt.show()
````
